```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
new_df = df[df['target'] != 0][['sepal width (cm)', 'petal length (cm)','target']]
```

```python
new_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['sepal width (cm)'],df['petal length (cm)'],c=df['target'],cmap='winter')
```

```python
df_train = new_df.iloc[:60,:].sample(10)
df_train
```

```python
# Taking only 10 rows for training
df = df.sample(100)
df_train = new_df.iloc[:60,:].sample(10)
df_val = new_df.iloc[60:80,:].sample(5)
df_test = new_df.iloc[80:,:].sample(5)
```

```python
df_train
```

```python
df_val
```

```python
df_test
```

```python
X_test = df_val.iloc[:,0:2].values
y_test = df_val.iloc[:,-1].values
```

## Case 1 Bagging

```python
# Data for Tree 1
df_bag = df_train.sample(8,replace=True)

X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

df_bag
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
dt_bag1 = DecisionTreeClassifier()
```

```python
## evaluation fuction
def evaluate(clf,X,y):
    clf.fit(X,y)
    plot_tree(clf)
    plt.show()
    plot_decision_regions(X.values, y.values, clf=clf, legend=2)
    y_pred = clf.predict(X_test)
    print(accuracy_score(y_test,y_pred))
```

```python
evaluate(dt_bag1,X,y)
```

```python
# Data for Tree 2
df_bag = df_train.sample(8,replace=True)

# Fetch X and y
X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

# print df_bag
df_bag
```

```python
dt_bag2 = DecisionTreeClassifier()
evaluate(dt_bag2,X,y)
```

```python
# Data for Tree 3
df_bag = df_train.sample(8,replace=True)

# Fetch X and y
X = df_bag.iloc[:,0:2]
y = df_bag.iloc[:,-1]

# print df_bag
df_bag
```

```python
dt_bag3 = DecisionTreeClassifier()
evaluate(dt_bag3,X,y)
```
