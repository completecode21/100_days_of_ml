```python
import pandas as pd
import numpy as np
import datetime as dt
import time
```

## Working WIth Date

```python
#convert the date object into datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#extract year
date['date_year'] = date['date'].dt.year
date
```

```python
# Extract month
date['date_month'] = date['date'].dt.month
date
```

```python
#exteact month name
date['date_month_name'] = date['date'].dt.month_name()
date
```

```python
#Extract day
date['date_day'] = date['date'].dt.day
date
```

```python
#Extract day name
date['date_day_name'] = date['date'].dt.day_name()
date
```

```python
#extract day of week 
date['date_day_od_week'] = date['date'].dt.dayofweek
date
```

```python
# Extract is weekend ?
date['date_weekend'] = np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
date
```

```python
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
date.head(4)
```

```python
#Extract the week no of the year
date['date_week_of_year'] = date['date'].dt.isocalendar().week
date
```

```python
#Extract quarter
date['date_quarter'] = date['date'].dt.quarter
date
```

```python
date['date_semester'] = np.where(date['date_quarter'].isin([1,2]),1,2)
date
```

```python
#Extract time elasped between date
import datetime

today = datetime.datetime.today()
today
```

```python
#elasped btwn date
today - date['date']
```

```python
#elasped btwn days
(today - date['date']).dt.days
```

```python
#month passed
np.round((today - date['date']) /np.timedelta64(1,'M'),0)
```

```python
# hour passed
np.round((today - date['date']) /np.timedelta64(1,'h'),0)
```

```python
# minute passed
np.round((today - date['date']) /np.timedelta64(1,'m'),0)
```

```python
#second passed
np.round((today - date['date']) /np.timedelta64(1,'s'),0)
```
