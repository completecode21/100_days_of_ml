```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
flight = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
#scatter plot (numerical-numerical)
sns.scatterplot(x = tips['total_bill'],y = tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
#bar plot (numerical-categorical) x = categorical , y = numerical
sns.barplot(x = titanic['Pclass'],y = titanic['Fare'],hue=titanic['Sex'])
```

```python
#Box plot (numerical-categrical) x = categorical, y = numerical
sns.boxplot(x = titanic['Sex'],y= titanic['Age'],hue=titanic['Survived'])
```

```python
# Distplot (numerical-categorical) x=categorical,y=numerical
sns.displot(titanic[titanic['Survived']==0]['Age'],kde=True)
sns.displot(titanic[titanic['Survived']==1]['Age'],kde=True)
```

```python
# HeatMap(categorical-categorical)
print(pd.crosstab(titanic['Pclass'],titanic['Survived']))
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
#ClusterMap(categorical-categorical)
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
